```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (20, 10)
plt.style.use('fivethirtyeight')
```

```python
head_brain_df = pd.read_csv('./Exercise_Problems/data/headbrain.csv')
print(head_brain_df.shape)
head_brain_df.head()
```

```python
x_head_size = head_brain_df['Head Size(cm^3)'].values
y_brain_weight = head_brain_df['Brain Weight(grams)'].values

x_mean = np.mean(x_head_size)
y_mean = np.mean(y_brain_weight)
total_length = len(x_head_size)
```

```python
numerator, denominator = 0, 0
for i in range(total_length):
    numerator += (x_head_size[i] - x_mean) * (y_brain_weight[i] - y_mean)
    denominator += (x_head_size[i] - x_mean)**2
beta_1 = numerator / denominator
beta_0 = y_mean - (beta_1 * x_mean)
print(f"Intercept is {beta_0} and Slope is {beta_1}")
```

```python
x_max = np.max(x_head_size) + 100
x_min = np.min(x_head_size) - 100
x = np.linspace(x_min, x_max, 1000)
y_hat = beta_0 + (beta_1 * x) # Create the line for plotting
```

```python
plt.scatter(x_head_size, y_brain_weight, label='Data Points')
plt.plot(x, y_hat, color='red', label='Regression Line')
# plt.plot(x_list[1], y_list[1], color='yellow')
plt.title('Head Size Vs Brain Weight with a Regression Line', loc='left')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.legend();
```
